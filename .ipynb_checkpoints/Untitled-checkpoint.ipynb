# Step 1: Load the product dataset

```python
# Load the products table

import pandas as pd
import glob

products = pd.read_csv("archive/product_info.csv")
print("Products shape:", products.shape)
products.head()
```

```python
# Grab all review CSVs into one file reviews.

review_files = glob.glob("archive/reviews_*.csv")

reviews = pd.concat([
    pd.read_csv(f,dtype={"Product_id":str},low_memory=False)
    for f in review_files],
                   ignore_index=True)
```

# Step 1. Making the data easy-to-use for the Decision Tree

```python
#price range
products["price_range"] = pd.cut(
    products["price_usd"],
    bins=[0,20,50, products["price_usd"].max()],
    labels=["Low","Medium","High"]
)

#rating
# Use qcut to split into 3 categories by actual distribution
products["rating_category"] = pd.qcut(products["rating"], q=3, labels=["Poor","Good","Excellent"])

#review
products["review_volume"] =pd.cut(
    products["reviews"],
    bins=[0,10,50, products["reviews"].max()],
    labels=["Low","Medium","High"]
)

#popularity
products["popularity"] =pd.cut(
    products["loves_count"],
    bins=[-1,3000,8000,products["loves_count"].max()],
    labels=["Low","Medium","High"]
)
```

# Step 3: Making BestSeller Target

```python

def bestseller(row):
    if row["rating_category"] == "Excellent" and row["review_volume"] == "High" and row["popularity"] == "High":
        return "High"
    elif row["rating_category"] == "Good" and row["review_volume"] == "Medium" and row["popularity"] == "Medium":
        return "Medium"
    else: 
        return "Low"

products["bestseller_potential"] =products.apply(bestseller, axis=1)
```

```python
print(products[["product_name","rating","reviews","loves_count","bestseller_potential"]].head(50))
      
```

# Step 4: Encode categorical features:

```python
from sklearn.preprocessing import LabelEncoder

le_rating =LabelEncoder()
le_review = LabelEncoder()
le_pop= LabelEncoder()

products["rating_enc"] =le_rating.fit_transform(products["rating_category"])
products["review_enc"] = le_review.fit_transform(products["review_volume"])
products["pop_enc"] = le_pop.fit_transform(products["popularity"])
```

# Step 5: Preparing X and y

```python
features = ["rating_enc","review_enc","pop_enc"]
X = products[features]
y = products["bestseller_potential"]

#encode target
le_target =LabelEncoder()
y_enc = le_target.fit_transform(y)
```

# Step 6: Training Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 

X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size= 0.2, random_state=42)

clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)

y_predict = clf.predict(X_test)
print(classification_report(y_test, y_predict, target_names=le_target.classes_))
```

# Step 7: Plotting the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(25,12))

plot_tree(
    clf, 
    feature_names=features,
    class_names=le_target.classes_,
    filled=True,
    rounded=True,
    fontsize=12
)

plt.show()
```

# Step 8: Pick the top 5 products with the highest predicted potential

```python
X_all = products[["rating_enc","review_enc","pop_enc"]].copy()
X_all = X_all.astype(int)

#predict
products['predicted_potential'] = clf.predict(X_all)
products['predicted_potential'] = le_target.inverse_transform(products['predicted_potential'])
```

```python
# High potential
high_bestsellers = products[products['predicted_potential'] == "High"]

# Sorting by popularity (or rating/reviews/loves_count) descending
top5_bestsellers = high_bestsellers.sort_values(
    by=['loves_count','reviews','rating'], 
    ascending=False
).head(5)

top5_bestsellers[['product_name','rating','reviews','loves_count','predicted_potential']]
```
